```python
import numpy as np
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
```

```python
%watermark --iversions
```

```python
X_num_train = np.load('../input/higgs_small_roc/X_num_train.npy')
X_num_val = np.load('../input/higgs_small_roc/X_num_val.npy')
X_num_test = np.load('../input/higgs_small_roc/X_num_test.npy')

y_train = np.load('../input/higgs_small_roc/y_train.npy')
y_val = np.load('../input/higgs_small_roc/y_val.npy')
y_test = np.load('../input/higgs_small_roc/y_test.npy')
```

```python
X_num_train.shape
```

```python
X_num_val.shape
```

```python
X_num_train_subsample, _, y_train_subsample, _ = train_test_split(

X_num_train, y_train, train_size=10_000, random_state=42)

X_num_val_subsample, _, y_val_subsample, _ = train_test_split(

X_num_val, y_val, train_size=10_000, random_state=42)

X_num_test_subsample, _, y_test_subsample, _ = train_test_split(

X_num_test, y_test, train_size=10_000, random_state=42)
```

```python
X_num_train_subsample.shape
```

```python
clf = XGBClassifier()
```

```python
clf.fit(X_num_train_subsample, y_train_subsample)
```

```python
val_preds = clf.predict_proba(X_num_val_subsample)[:,1]
test_preds = clf.predict_proba(X_num_test_subsample)[:,1]
```

```python
val_preds
```

```python
roc_auc_score(y_val_subsample, val_preds)
```

```python
roc_auc_score(y_test_subsample, test_preds)
```
