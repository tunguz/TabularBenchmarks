```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
import pandas as pd
```

```python
train_data = TabularDataset('../input/train.csv.gz')
```

```python
train_data.head()
```

```python
%%time
predictor = TabularPredictor(label='label').fit(train_data)
```

```python
test_data = TabularDataset('../input/test.csv.gz')
```

```python
y_pred = predictor.predict(test_data)
```

```python
y_pred
```

```python
predictor.evaluate(train_data, silent=True)
```

```python
predictor.leaderboard(train_data)
```

```python
sample_submission = pd.read_csv('../input/sample_submission.csv.gz')
sample_submission.head()
```

```python
sample_submission.label = y_pred
sample_submission.to_csv('../output/Autogluon0.csv', index=False)
sample_submission.head()
```
